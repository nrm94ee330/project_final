```python

hp_df_2015 = pd.read_excel(file_2015, sheet_name="Figure2.2",skiprows=3)
#print(hp_df_2015.columns)

hp_df_2015=hp_df_2015.drop(hp_df_2015.columns.to_series()["Trust in national government":"Most people can be trusted, WVS round 2010-2014"], axis=1)
hp_df_2015=hp_df_2015.drop(["WP5 Country"],axis=1)

hp_df_2015=hp_df_2015.rename(columns={"Dystopia + residual":"Dystopia residual","Explained by: Log GDP per capita":"Explained by: GDP per capita","country":"Country"})

hp_df_2015["Happiness score"]=hp_df_2015["Explained by: GDP per capita"]+ hp_df_2015["Explained by: Social support"] + hp_df_2015["Explained by: Healthy life expectancy"]+ hp_df_2015["Explained by: Freedom to make life choices"]+ hp_df_2015["Explained by: Generosity"] + hp_df_2015["Explained by: Perceptions of corruption"] + hp_df_2015["Dystopia residual"]

hp_df_2015=hp_df_2015.drop(hp_df_2015.columns.to_series()["Regional indicator":"Ladder score in Dystopia"], axis=1)

hp_df_2015["Happiness.Rank"]=hp_df_2015["Happiness score"].rank(ascending=0)

hp_df_2015["year"]=2014
hp_df_2015.head()
```

```python

hp_df_2016 = pd.read_excel(file_2016, sheet_name="Figure2.2")
hp_df_2016=hp_df_2016.rename(columns={"Dystopia (2.33) + residual":"Dystopia residual"})
hp_df_2016=hp_df_2016.dropna(subset=['Country'])
hp_df_2016["Happiness.Rank"]=hp_df_2016["Happiness score"].rank(ascending=0)
hp_df_2016["year"]=2015
hp_df_2016=hp_df_2016.drop(hp_df_2016.columns.to_series()["Whisker-high":"Whisker-low"], axis=1)
hp_df_2016=hp_df_2016.drop(hp_df_2016.columns.to_series()["Unnamed: 11":"Figure 2.2: Ranking of Happiness 2013-2015"], axis=1)
hp_df_2016.head()

```

```python

hp_df_2017 = pd.read_excel(file_2017, sheet_name="Figure2.2")
hp_df_2017=hp_df_2017.rename(columns={"Dystopia (1.85) + residual":"Dystopia residual"})
hp_df_2017=hp_df_2017.drop(hp_df_2017.columns.to_series()["Whisker-high":"Whisker-low"], axis=1)
hp_df_2017=hp_df_2017.dropna(subset=['Country'])
hp_df_2017["Happiness.Rank"]=hp_df_2017["Happiness score"].rank(ascending=0)
hp_df_2017["year"]=2016
hp_df_2017.head()
```

```python
hp_df_2018 = pd.read_excel(file_2018, sheet_name="Figure2.2")
hp_df_2018=hp_df_2018.rename(columns={"Dystopia (1.92) + residual":"Dystopia residual"})

hp_df_2018=hp_df_2018.drop(hp_df_2018.columns.to_series()["Whisker-high":"Whisker-low"], axis=1)
hp_df_2018=hp_df_2018.drop(hp_df_2018.columns.to_series()["Unnamed: 11":"Unnamed: 23"], axis=1)
hp_df_2018["Happiness.Rank"]=hp_df_2018["Happiness score"].rank(ascending=0)
hp_df_2018["year"]=2017
hp_df_2018=hp_df_2018.dropna(subset=['Country'])
hp_df_2018.head()
```
